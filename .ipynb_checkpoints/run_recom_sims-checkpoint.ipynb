# Create Initial District Assignment

#### Load modules

```python
import os
import geopandas as gpd
import pandas as pd
import numpy as np
import random
import math
import gerrychain
from gerrychain import (GeographicPartition, Partition, Graph, MarkovChain,
                        proposals, updaters, constraints, accept, Election)
from gerrychain import MarkovChain, accept
from gerrychain.updaters import Tally, cut_edges
from gerrychain.tree import recursive_tree_part
from gerrychain.constraints import contiguous, Validator
from gerrychain.proposals import recom
from gerrychain.metrics import efficiency_gap, mean_median
from functools import partial
```

#### Read in shapefile and clean up column names

```python
# Load shapefile
gdf = gpd.read_file('Output/test/test.shp')

# Ensure the GeoDataFrame has a unique index
gdf = gdf.reset_index(drop = True)

# Rename variables
gdf = gdf.rename(columns = {"prcnct_": "precinct_id", "popultn": "population", 
                          "pr_mnrt": "per_minority", "dm_vtsh": "dem_voteshare"})

# Calculate bvap
gdf['bvap'] = np.round(gdf['per_minority'] * gdf['population']).astype(int)

# Calculate wvap
gdf['wvap'] = gdf['population'] - gdf['bvap']

# Calculate rep_voteshare
gdf['rep_voteshare'] = 1 - gdf['dem_voteshare']
```

```python
gdf['precinct_id']
```

#### Create graph object

```python
# Create graph from GeoDataFrame
graph = Graph.from_geodataframe(gdf)
```

```python
# Add population data to graph
for node in graph.nodes():
    graph.nodes[node]["population"] = gdf.loc[node, "population"]
    graph.nodes[node]['dem_voteshare'] = gdf.loc[node, 'dem_voteshare']
```

```python
# Calculate the ideal population for each district
total_population = sum(gdf['population'])
target_CD_population = 750000
num_rows = len(gdf)
num_districts = math.ceil(total_population / target_CD_population)
ideal_population = total_population / num_districts
```

```python
# Set up Election
election = Election("ELECTION", {"Dem": "dem_voteshare", "Rep": 'rep_voteshare'})
```

```python
my_updaters = {
    'population': Tally('population'),
    'cut_edges': cut_edges,
    'ELECTION': election
}
```

```python
# Create initial partition using from_random_assignment
initial_partition = Partition.from_random_assignment(
    graph, 
    n_parts=num_districts, 
    epsilon=0.05,  # 5% population deviation
    pop_col="population",
    updaters=my_updaters
)
```

### Plot initial assignment

```python
%matplotlib inline
import matplotlib.pyplot as plt

initial_partition.plot(
    gdf,
    figsize=(8, 8),
    cmap='tab20',  # color map for districts
    edgecolor='black',
    linewidth=0.5
)
plt.title("Initial CD Partition")
plt.axis('off')
plt.show()
```

```python
# Print out the population of each district
for district, pop in initial_partition["population"].items():
    print(f"District {district} population: {pop}")
print(f"Ideal population: {ideal_population}")
```

```python
# Write initial district assignment back to data
gdf['init_CD'] = [initial_partition.assignment[node] for node in graph.nodes()]
```

# Run ReCom

```python
# Set up constraints
pop_constraint = constraints.within_percent_of_ideal_population(initial_partition, 0.05)
compactness_bound = constraints.UpperBound(
    lambda p: len(p["cut_edges"]),
    2*len(initial_partition["cut_edges"])
)
```

```python
# Set up proposal function
proposal = partial(recom,
                   pop_col="population",
                   pop_target=ideal_population,
                   epsilon=0.05,
                   node_repeats=2
                  )
```

```python
# Set up the chain
chain = MarkovChain(
    proposal=proposal,
    constraints=[
        constraints.contiguous,
        constraints.within_percent_of_ideal_population(initial_partition, 0.05), # or pop_constraint
        compactness_bound
    ],
    accept=accept.always_accept,
    initial_state=initial_partition,
    total_steps=10000
)
```

```python
# Run the chain
metrics = []
CD_assignments = []
for i, partition in enumerate(chain):
    if i % 100 == 0:
        print(f"Step {i}")
    metrics.append({
        "step": i,
        "cut_edges": len(partition["cut_edges"]),
        "efficiency_gap": efficiency_gap(partition["ELECTION"]),
        "mean_median": mean_median(partition["ELECTION"])
    })
    CD_assignments.append({
        'step': i,
        'district_id': [partition.assignment[node] for node in graph.nodes()]
    })
```

#### Write CD assignments back to data

```python
df_assignments = pd.DataFrame(CD_assignments)
```

```python
# array to hold all the district assignments
num_steps = len(df_assignments)
num_units = len(df_assignments['district_id'].iloc[0])
all_assignments = np.zeros((num_units, num_steps), dtype=int)

for i, row in enumerate(df_assignments['district_id']):
    all_assignments[:, i] = row
```

```python
# make dataframe and add precinct_id
df_pivoted = pd.DataFrame(all_assignments, columns=[f'step_{i+1}' for i in range(num_steps)])
df_pivoted['precinct_id'] = gdf['precinct_id']
df_pivoted['init_CD'] = gdf['init_CD']
```
